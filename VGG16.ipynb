**Description**:
VGG-16 is a convolutional neural network that is 16 layers deep.


```python
from keras.applications.vgg16 import VGG16
model = VGG16()
model.summary()
```

```python
from keras.preprocessing.image import load_img
img = load_img('/content/Tennis ball.jpg',target_size = (224,224))
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#img = mpimg.imread('/content/Tennis ball.jpg')
plt.imshow(img)
```

```python
img.shape
```

```python
import numpy as np
img = np.expand_dims(img,axis = 0)
img.shape
```

```python
from keras.applications.vgg16 import preprocess_input
img = preprocess_input(img)
img.shape
```

```python
prediction = model.predict(img)
```

```python
prediction.shape
```

```python
from tensorflow.keras.applications.mobilenet import decode_predictions
pred_classes = decode_predictions(prediction, top=5)
for i in pred_classes[0]:
    print(i)
```
